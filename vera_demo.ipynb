### emocess(fileToEmocess)

This function takes the path of an audio file as an input. 

Converts the file to MP3 if it is not already in that format. Then splits and sends the data to VERA service.

Returns a dataframe that has emovecs for rows.

```python
from vera.emocess import emocess
```

```python
fileToEmocess = 'data_samples/cropped_.m4a'
df = emocess(fileToEmocess)
df.head(10)
```

### plotEmovec

Plot the given emovec

```python
from vera.emocess import plotEmovec

f = plotEmovec(df)
f.savefig('out_fig.png')
```
